# Check wheter given number is leap year or not
1. divisble by 4 but not divisible by 100
2. divisible by 400

```python
year = int(input("Please enter the year : "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(f"Yes, {year} is a leap year")
else:
    print(f"{year} is non-leap year")
```

# Calculate BMI catgory

```python
weight = float(input("Enter weight in kg : "))
height = float(input("Enter height in m : "))

bmi = weight / height**2

print(f"BMI : {bmi:.2f} kg/m^2")

if bmi < 18.5:
    print("Underweight")
elif bmi >= 18.5 and bmi < 25:
    print("Normal Weight")
elif bmi >= 25 and bmi < 30:
    print("OverWeight")
elif bmi >= 30 and bmi < 35:
    print("Obesitity Class 1")
elif bmi >= 35 and bmi < 40:
    print("Obesitity Class 2")
else:
    print("Obesity Class 3")
```
